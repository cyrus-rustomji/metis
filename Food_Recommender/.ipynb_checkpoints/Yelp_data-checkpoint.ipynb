# Yelp Data Clean

```python
business = pd.read_csv('Data/yelp_business.csv')[['business_id','name','address','city','state','postal_code','stars','review_count','is_open','categories']]
reviews = pd.read_csv('Data/yelp_review.csv')[['review_id','business_id','date','text']]
hours = pd.read_csv('Data/yelp_business_hours.csv')
business.head()
```

```python
# use toronto because in vegas people eat out
business.city.value_counts().head()
```

```python
business = business[business.city == 'Toronto']
business_reviews = business.merge(reviews, left_on='business_id', right_on='business_id', how='left')
combine = business_reviews.merge(hours, left_on='business_id', right_on='business_id', how='left')
```
