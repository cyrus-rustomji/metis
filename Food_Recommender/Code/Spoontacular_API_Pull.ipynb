# Put all recipes in one dictionary

```python
spoon = requests.get("https://spoonacular-recipe-food-nutrition-v1.p.mashape.com/recipes/informationBulk?ids=690978%2C720070%2C428467%2C574946%2C473511%2C728903%2C495417%2C244482%2C599841",
  headers={
    "X-Mashape-Key": "1lgfQw0SXHmshZG3AhwpBY3kw2U6p1ACeknjsnqs9aKB54jT36",
    "X-Mashape-Host": "spoonacular-recipe-food-nutrition-v1.p.mashape.com"
  }
)
final = spoon.json()
final
```

```python
# find names of recipes
i = 0
z = 0
q = 0
d = 0
r = 0
recipes = []
for x in final:
    if i == 9 or z == 9 or q == 9 or r == 9:
        break
    header = []
    for x in final[i]['title']:
        header = (final[i]['title'])

    # ingredients dictionary
    lst = []
    lst2 = []
    ingredients = {}
    for a in final[z]['extendedIngredients']:
        for key,value in a.items():
            lst.append(a['original'])
    for item in lst:
        if item not in lst2:
            lst2.append(item)

    ingredients['Ingredients'] = lst2

    # instructions dictionary
    duplicate_instrc = []
    instrc = []
    dictionary = {}
    instructions = {}
    if len(final[q]['analyzedInstructions']) == 0:
        instrc = []
    else:
        for r in final[q]['analyzedInstructions'][0]['steps']:
            for key,value in r.items():
                duplicate_instrc.append(r['step'])
        for items in duplicate_instrc:
            if items not in instrc:
                instrc.append(items)

    lst3 = []
    lst4 = []
    lst5 = []
    if len(final[i]['analyzedInstructions']) == 0:
        lst5 = []
    else:
        for y in final[i]['analyzedInstructions'][0]['steps']:
            for key,value in y.items():
                lst3.append(y['number'])
        for elements in lst3:
            if elements not in lst5:
                lst5.append(elements)
        for item in lst:
            if item not in lst4:
                lst4.append(item)

    dictionary = dict(zip(lst5, instrc))
    dictionary
    instructions['Instructions'] = dictionary



    # dish types dictionary -- complete
    dish_types = {}
    ready = {}
    diets = {}
    wine = {}
    cuisines = {}
    full_dictionary = {}
    vegetarian = {}
    vegan = {}
    gluten_free = {}
    dairy_free = {}
    dish_types['Dish Types'] = final[i]['dishTypes']

    ready['Time to Cook'] = final[i]['readyInMinutes']

    # diets dictionary
    diets['Diets'] = final[i]['diets']

    # wine pairings dictionary
    wine['Wine Pairings'] = final[i]['winePairing']['pairedWines']

    # cuisines dictionary
    cuisines['Cuisines'] = final[i]['cuisines']
    vegetarian['Vegetarian'] = final[i]['vegetarian']
    vegan['Vegan'] = final[i]['vegan']
    gluten_free['Gluten-Free'] = final[i]['glutenFree']
    dairy_free['Dairy-Free'] = final[i]['dairyFree']

    # full dictionary
    title = final[i]['title']
        
    full_dictionary[title] = [instructions,ingredients,ready,dish_types,diets,cuisines,vegetarian,vegan,gluten_free,dairy_free,wine]

    # recipes dictionary
    recipes.append(full_dictionary)
    i+=1
    z+=1
    q+=1
    d+=1

# full_dictionary
recipes
# ready
# dish_types
# diets
# wine
# cuisines
# full_dictionary
# dish_types
# instructions
# instrc
# ingredients
# header
```
